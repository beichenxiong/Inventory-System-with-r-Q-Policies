```python
lamb=[]
L=[]
h=[]
p=[]
k=[]
s=[]
with open(r'E:\121.csv', 'r') as csvfile: 
    reader = csv.reader(csvfile)
    for line in reader:
        lamb.append(float(line[2]))
        L.append(float(line[4]))
        h.append(float(line[5]))
        p.append(float(line[6]))
        k.append(float(line[3]))
        s.append(float(line[1]))
```

```python
def Algorithm1(lamb,L,h,p):
    Alpha=p/(p+h)
    mean=lamb*L
    y=st.poisson.ppf(Alpha,mean)+1
    return y
def Algorithm2(qmin,qmax,lamb,k,L,h,p):
    G1=0
    G2=0
    c=0
    while (min(G1,G2)<=c):
        G=0
        Ps1=0
        Ps2=0
        r=qmin-1
        Q=qmax-r
        for i in range(0,r):
            Ps1=Ps1+st.poisson.cdf(i,lamb*L)
        for i in range(0,qmax+1):
            Ps2=Ps2+st.poisson.cdf(i,lamb*L)
        G1=(h+p)*Ps1+p*(lamb*L+1-qmin)
        G2=(h+p)*Ps2+p*(lamb*L-1-qmax)
        y=qmin
        while (y<=qmax):
            Ps=0
            for i in range(0,y):
                Ps=Ps+st.poisson.cdf(i,lamb*L)
            G=G+(h+p)*Ps+p*(lamb*L-y)
            y=y+1
        c=(lamb*k+G)/Q
        if G1<=G2:
            qmin=qmin-1
        else:
            qmax=qmax+1
    return c,r,Q
def Algorithm3(r,Q,lamb,k,L,h,p):
    y1=r-1+1
    G1=0
    G2=0
    y2=r+1
    c1=0
    c2=0
    while (y1<=r-1+Q):
        Ps1=0
        for i in range(0,y1):
            Ps1=Ps1+st.poisson.cdf(i,lamb*L)
        G1=G1+(p+h)*Ps1+p*(lamb*L-y1)
        y1=y1+1
    c1=(lamb*k+G1)/Q
    while (y2<=r+Q-1):
        Ps2=0
        for i in range(0,y2):
            Ps2=Ps2+st.poisson.cdf(i,lamb*L)
        G2=G2+(p+h)*Ps2+p*(lamb*L-y2)
        y2=y2+1
    c2=(lamb*k+G2)/(Q-1)
    if c1<=c2:
        r=r-1
        return c1,r,Q
    else:
        Q=Q-1
        return c2,r,Q
```

```python
y1=[]
r1=[]
Q1=[]
c1=[]
for i in range(0,30):
    lamb1=lamb[i]
    L1=L[i]
    h1=h[i]
    p1=p[i]
    y=Algorithm1(lamb1,L1,h1,p1)
    y1.append(y)
print y1
for i in range(0,30):
    qmin=int(y1[i])
    qmax=int(y1[i])
    k1=k[i]
    lamb2=lamb[i]
    L2=L[i]
    h2=h[i]
    p2=p[i]
    y8=Algorithm2(qmin,qmax,lamb2,k1,L2,h2,p2)
    c1.append(y8[0])
    r1.append(y8[1])
    Q1.append(y8[2])
print c1,r1,Q1
```

```python
c2=[]
r2=[]
Q2=[]
for i in range(0,30):
    r=r1[i]
    Q=Q1[i]
    c=c1[i]
    k2=k[i]
    lamb3=lamb[i]
    L3=L[i]
    h3=h[i]
    p3=p[i]
    y=Algorithm3(r,Q,lamb3,k2,L3,h3,p3)
    c2.append(y[0])
    r2.append(y[1])
    Q2.append(y[2])
print c2,r2,Q2
```
